# student mark analyzer


```python
import numpy as np
```

```python
marks = np.array([67, 89, 45, 90, 76, 55, 32])
```

```python
print("students marks")
```

```python
print(marks)
```

```python
print("average marks:", marks.mean())
```

```python
print("highes marks:", marks.max())
```

```python
print("lower marks:", marks.min())
```

```python
passed = marks[marks >= 50]
```

```python
print("passed students marks:")
```

```python
print(passed)
```

```python
failed = marks[marks <= 50]
```

```python
print("failed student marks:")
```

```python
print(failed)
```

```python
percentage = (marks/100)*100
```

```python
print("percentages:")
```

```python
print(percentage)
```

```python
for i in marks:
    if i >= 90:
        print("grade A")
    if i >= 75:
        print("grade B")
    if i >= 50:
        print("grade C")
    else:
        print("fail")
```

```python
marks = [67, 89, 45, 90, 76, 55, 32]
pass_count = 0
fail_count = 0
for i in marks:
    if i >= 50:
        print("pass")
        pass_count += 1
    else:
        print("fail")
        fail_count += 1
print("passed students:", pass_count)
print("failed studets:", fail_count)
```
